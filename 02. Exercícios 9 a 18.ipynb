## Exercício 10

Faça um programa que peça ao usuário o raio de um círculo e exiba na tela a área e o perímetro desse círculo (considere pi = 3.14). Aproxime para 2 casas decimais

- Perímetro: **P = 2.π . r**
- Área: **A = π . r^2**

```python
raio = float(input("Digite o valor do raio do círculo: "))
```

```python
numero_pi = 3.14
perimetro = 2 * numero_pi * raio
area = numero_pi * pow(raio, 2)
```

```python
print(f"Área do circulo: {area:.2f} \nPerímetro do circulo: {perimetro:.2f}")
```

## Exercício 12

Faça um programa que leia a temperatura em graus Celsius e exiba a temperatura em graus Fahrenheit. A fórmula para converter de Celsius para Fahrenheit é: F = (9/5)\*C + 32

```python
temp_celsius = float(input("Digite a temperatura em graus Celsius: "))
```

```python
fahrenheit = ((9/5) * temp_celsius) + 32
```

```python
print(f"A temperatura {temp_celsius} C° é igual a {fahrenheit:.2f} F°")
```

## Exercício 14

Faça um programa que leia o peso e a altura de uma pessoa e exiba o índice de massa corporal (IMC) dela. A fórmula para calcular o IMC é: IMC = peso/altura², com aproximação em 3 casas decimais.

```python
peso = float(input("Digite seu peso: "))
altura = float(input("Digite sua altura: "))
```

```python
imc = peso / pow(altura, 2)
```

```python
print(f"Indice de Massa Corporal (IMC): {imc:.3f}")
```

## Exercício 17

Escreva um programa Python que leia um número e retorne seu quadrado, sua raíz quadrada e sua raíz cúbica, com aproximação em 2 casas decimais.

```python
num = float(input("Digite um número: "))
```

```python
quadrado = num ** 2
raiz_quadrada = num ** 0.5 # ou num ** 1/2
raiz_cubica = num ** (1/3) # ou num ** 1/3
```

```python
print(f"{num} elevado ao quadrado = {quadrado:.2f}")
print(f"Raiz quadrada de {num} = {raiz_quadrada:.2f}")
print(f"Raiz cúbica de {num} = {raiz_cubica:.2f}")
```

## Exercício 18

Escreva um programa Python que leia o valor de dois catetos e retorne o valor da hipotenusa, assumindo que seja possível formar um triângulo

```python
cateto_a = float(input("Digite o valor do cateto a: "))
cateto_b = float(input("Digite o valor do cateto b: "))
```

```python
hipotenusa = ((cateto_a ** 2) + (cateto_b ** 2)) ** 0.5
```

```python
print(f"O valor da hipotenusa é igual a: {hipotenusa}")
```
