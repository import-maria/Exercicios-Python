## Exercício 36

Faça um programa que solicite ao usuário digitar uma frase e conte quantas vogais existem na frase utilizando um loop "for".

```python
frase = input("Digite uma frase: ").lower()
```

```python
quantidade_de_vogais = 0
vogais = ['a', 'e', 'i', 'o', 'u']

for letra in frase:
  for vogal in vogais:
    if letra == vogal:
      quantidade_de_vogais += 1
```

```python
print(f"Quantidade de vogais na frase '{frase}' = {quantidade_de_vogais}")
```

## Exercício 38

Faça um programa que imprima os números de 1 a 100, mas substitua os múltiplos de 3 pela palavra "Fizz", os múltiplos de 5 pela palavra "Buzz" e os múltiplos de 3 e 5 pela palavra "Fizz Buzz".

```python
lista = []
```

```python
for num in range(1, 100, 1):
  if (num % 5 == 0) and (num % 3 == 0):
    lista.append("Fizz Buzz")

  elif num % 5 == 0:
    lista.append("Buzz")

  elif num % 3 == 0:
    lista.append("Fizz")

  else:
    lista.append(num)
```

```python
print(lista)
```

## Exercício 39

Faça um programa que solicite ao usuário digitar uma lista de números e encontre o maior número da lista.

```python
continuar = 'y'
lista = []
```

```python
while continuar == 'y':
  numero = float(input("Digite um número: "))
  lista.append(numero)
  continuar = input("Deseja continuar adicionando números a lista? [y/n] ")
```

```python
print(lista)
```

```python
# encontrar o maior número da lista
maior_numero = 0

for i in range(0, len(lista)):
  numero = lista[i]
  if numero > maior_numero:
    maior_numero = numero
    print(maior_numero)
  else:
    continue
```

```python
print(maior_numero)
```

## Exercício 40

Faça um programa que solicite ao usuário digitar uma palavra e verifique se a palavra é um palíndromo utilizando um loop "for".

```python
palavra = input("Digite uma palavra: ")
```

```python
def validar_palíndromo(palavra):
  palava_invertida = palavra[::-1]
  indice = 0
  contador = 0

  for letra in palavra:
    if letra == palava_invertida[indice]:
      contador += 1
    else:
      continue
    indice += 1

  print(f"A palavra é um palíndromo? {indice == len(palavra)}")
```

```python
validar_palíndromo(palavra)
```

```python
palavra = "Maria"
```

```python
validar_palíndromo(palavra)
```

## Exercício 41

Faça um programa que solicite ao usuário digitar um número e, em seguida, calcule e imprima o fatorial desse número.

```python
numero8 = int(input("Digite um número: "))
```

```python
fatorial = numero8

for i in range(1, numero8):
  fatorial *= i
  print(fatorial)
```

```python
print(f"O fatorial do número {numero8} é = {fatorial}")
```
