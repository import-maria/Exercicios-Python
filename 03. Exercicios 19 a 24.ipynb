## Exercício 19

Faça um programa que receba uma frase do usuário e retorne: (i) a quantidade de caracteres da frase (ii) a quantidade de letras 'e' na frase (iii) o índice da primeira substring 'ou' (iv) a frase toda maiúscula (v) a frase toda minúscula.

```python
frase = input("Digite uma frase: ")
```

(i) a quantidade de caracteres da frase

```python
print(f'Quantidade de caracteres da frase: {len(frase)}')
```

(ii) a quantidade de letras 'e' na frase

**Observe que o 'é' não será considerado:**

```python
print(f"Quantidade de 'e' na frase: {frase.count('e')}")
```

(iii) o índice da primeira substring

```python
print(f"Indice da primeira substring: {frase.find('como')}")
```

(iv) a frase toda maiúscula

```python
print(f"Frase com todas as letras maiúsculas: {frase.upper()}")
```

(v) a frase toda minúscula.

```python
print(f"Frase com todas as letras minúscula: {frase.lower()}")
```
