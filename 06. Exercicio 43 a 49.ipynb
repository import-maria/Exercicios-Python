## Exercício 43

Faça um programa que receba números do usuário até ele digitar -1 e armazene todos esses números em uma lista. Por fim, exiba na tela: "A soma dos números que você passou foi: {soma}".

```python
def somar_lista():
  lista = []
  n = float(input("Digite um número: "))

  while n != -1:
    lista.append(n)
    n = float(input("Digite um número: "))

  print(lista)

  return sum(lista)
```

```python
soma = somar_lista()
print(f"A soma de todos os elementos da lista é = {soma}")
```

## Exercício 44

Faça um programa que receba informações de 4 pessoas. As informações são: "nome", "idade" e "cidade". Armazene essas informações em uma lista de dicionários e, ao final, retorne a seguinte mensagem: "{nome} é o mais velho e mora em {cidade}."

```python
def quem_e_o_mais_velho():
  lista = []
  mais_velho = {}
  maior_idade = 0

  for i in range(0, 4, 1):
    nome = input(f"Digite o nome da pessoa {i}: ")
    idade = int(input(f"Digite a idade da pessoa {i}: "))
    cidade = input(f"Digite a cidade da pessoa {i}: ")

    dicionario = dict(nome= nome, idade= idade, cidade= cidade)

    lista.append(dicionario)

  print(lista)

  for i in lista:
    if i['idade'] > maior_idade:
      mais_velho = i
    else:
      continue

  print(f"{mais_velho['nome']} é o mais velho e mora em {mais_velho['cidade']}")
```

```python
print(quem_e_o_mais_velho())
```

Outra forma é utilizando a função sorted para ordernar:

```python
# utilizar essa função para encontrar dentro da lista, o valor idade de cada dicionário
def ordenar(elemento):
    return elemento['idade']
```

```python
pessoas = [{'nome': 'Maria', 'idade': 20, 'cidade': 'SP'}, {'nome': 'Laissa', 'idade': 18, 'cidade': 'ZL'}, {'nome': 'Adriana', 'idade': 49, 'cidade': 'PE'}, {'nome': 'Daniel', 'idade': 55, 'cidade': 'Bragança Paulista'}]
```

```python
lista_ordenada = sorted(pessoas, key = ordenar, reverse = True)

# lista fica ordenada em order decrescente pelo valor da idade
lista_ordenada
```

```python
# Podemos também utilizar um lambda (uma função diretamente nos parenteses)
lista_ordenada = sorted(pessoas, key = lambda elementos: elementos['idade'], reverse = True)

lista_ordenada
```

```python
# ou utilizando a função itemgetter
from operator import itemgetter

lista_ordenada = sorted(pessoas, key = itemgetter('idade'), reverse = True)

lista_ordenada
```

## Exercício 45

Faça um programa que peça ao usuário para digitar alguns números até ele digitar -1. Armazene todos esses números em uma lista e, por fim, exiba na tela essa lista com todos os elementos somados em 10 unidades (utilize a função **map()**).

```python
def criar_lista():
  lista = []

  while n != -1:
    lista.append(n)
    n = int(input("Digite um número: "))

  return lista
```

```python
lista = criar_lista()
```

```python
print(lista)
```

A função map permite expressar de forma concisa a operação que será aplicada a cada elemento da sequência, tornando o código mais legível e de fácil compreensão.

```python
multiplica_lista_por_10 = list(map(lambda numero: numero + 10, lista))

print(multiplica_lista_por_10)
```

## Exercício 46

Faça um programa que receba duas entradas: a quantidade de linhas e colunas, nessa ordem, de uma matriz. Após isso, solicite que ele digite o valor correspondente a cada posição dessa matriz: "Digite o valor para a linha {i} e coluna {j}: ". Ao final, retorne a multiplicação de todos os elementos dessa matriz.

```python
linhas = int(input("Digite a quantidade de linhas que a matriz deverá possuir: "))
colunas = int(input("Digite a quantidade de  colunas que a matriz deverá possuir: "))
```

```python
matriz = []

for linha in range(0, linhas):
  listas_da_matriz = [] # dessa forma remove os elementos anteriores da lista

  for coluna in range(0, colunas):
    valor = int(input(f"Digite o valor da linha {linha} coluna {coluna}: "))
    listas_da_matriz.append(valor)

  matriz.append(listas_da_matriz)
```

```python
print(matriz)
```

```python
# multiplicação de todos os elementos da matriz
multiplicador = 1

for linha, coluna in enumerate(matriz):
    for elemento in coluna:
        multiplicador *= elemento
```

```python
print(f"A multiplicação de todos os elementos dentro da matriz é = {multiplicador}")
```

## Exercício 48

Faça um programa que receba palavras de um usuário até que ele digite "goiabada" e adicione todas essas palavras em uma lista. Após isso, retorne uma lista com todas as palavras que não possua vogal (utilize a função **filter()**).

```python
def filtra_vogais(palavra):
    vogais = ['a', 'e', 'i', 'o', 'u']
    for vogal in vogais:
        if vogal in palavra:
            return False
    return True
```

```python
lista = []

palavras = input("Digite uma palavra (caso queira parar, digite goiaba): ").strip().lower()

while palavras != 'goiaba':
  palavras = input("Digite uma palavra (caso queira parar, digite goiaba): ").strip().lower()
  lista.append(palavras)

palavras_sem_vogal = list(filter(filtra_vogais, lista))
```

```python
print(f"Palavras da lista que não possuem vogais: {palavras_sem_vogal}")
```

## Exercício 49

Faça um programa que receba o nome e a temperatura máxima de várias cidades e ao longo de uma semana e armazene essas informações em um dicionário, onde a chave é o nome da cidade e o valor é a temperatura máxima (pare de receber informações quando o usuário digitar "goiabada" no nome da cidade). O programa deve retornar em quantos graus a maior temperatura supera a média das temperaturas.

- receba o nome e a temperatura máxima de várias cidades e ao longo de uma semana e armazene essas informações em um dicionário

```python
cidades = {}
cidade = input('Digite o nome da cidade (caso queira parar, digite chocolate): ')
```

```python
while cidade != 'chocolate':
  cidades[cidade ] = float(input(f"Digite a temperatura da cidade {cidade}: "))
  cidade = input('Digite o nome da cidade (caso queira parar, digite chocolate): ')
```

```python
media_temperaturas = sum(cidades.values()) / len(cidades)

print(f"Média das temperaturas: {media_temperaturas:.2f}")
```

```python
maior_temperatura = 0

for temperatura in cidades.values():
  if temperatura > maior_temperatura:
    maior_temperatura = temperatura
  else:
    continue

print(f"Em quantos graus a maior temperatura supera a média das temperaturas? R: {maior_temperatura - media_temperaturas:.2f}")
```
