## Exercício 74
Ordene o DataFrame pela coluna ***'highest_price_per_mwh'*** em ordem decrescente.

```python
hppm_gas
```

```python
df_gas_price.sort_values(by = ['highest_price_per_mwh', 'lowest_price_per_mwh'], ascending = False)
```

Quando usamos o comando `df_gas_price.sort_values(by=['highest_price_per_mwh', 'lowest_price_per_mwh'], ascending=False)`, os dados são ordenados primeiro pela 'highest_price_per_mwh' e, em seguida, pela 'lowest_price_per_mwh'. Isso significa que a ordenação ocorre de acordo com a primeira coluna especificada e, caso haja valores iguais nessa coluna, a ordenação subsequente será aplicada à segunda coluna especificada.

## Exercício 82
Realize um merge entre ***'gas_prices'*** e ***'train'*** usando a coluna ***'origin_date'*** e ***'datetime'***

```python
# criando uma coluna nova na tabela train só com a data do dia, sem as horas
df_train['date'] = df_train.datetime.dt.date

# o comando acima cria a coluna com o tipo de dado object, então novamente convertemos
df_train['date'] = pd.to_datetime(df_train['date'])
```

```python
# renomeamos a coluna origin_date para date na tabela gas_price
gas_renamed = df_gas_price.rename(columns = {'origin_date': 'date'})

# fazendo o merge - left join
merged = pd.merge(df_train, gas_renamed, on = 'date', how = 'left')
merged
```
