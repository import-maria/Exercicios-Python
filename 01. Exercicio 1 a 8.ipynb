## Exercício 6

Faça um programa que leia algo digitado pelo usuário, mostre seu tipo e tudo a respeito dele.


<ul>
    <li> isdigit(): Retorna True se todos os caracteres da string são dígitos, e False caso contrário.
    <li> isalpha(): Retorna True se todos os caracteres da string são letras do alfabeto, e False caso contrário.
    <li> isalnum(): Retorna True se todos os caracteres da string são alfanuméricos (letras ou dígitos), e False caso contrário.
    <li> islower(): Retorna True se todos os caracteres da string são letras minúsculas, e False caso contrário.
    <li> isupper(): Retorna True se todos os caracteres da string são letras maiúsculas, e False caso contrário.
    <li> isspace(): Retorna True se todos os caracteres da string são caracteres de espaço em branco, como espaço, tabulação ou quebra de linha, e False caso contrário.
</ul>

```python
entrada = input("Digite qualquer coisa que queira digitar :) - ")
```

```python
# verificar se possui apenas letras do alfabeto
entrada.isalpha()
```

```python
# verificar se todos os caracteres da entrada são números (0 a 9)
entrada.isdigit()
```

```python
so_numeros = input("Digite apenas números: ")
so_numeros.isdigit()
```

```python
# verificar se há letras junto com números, se tiver um caractere especial então ele irá retornar false
entrada.isalnum()
```

```python
# verifica se todos os caracteres são minusculos
entrada.islower()
```

```python
# verifica se todos os caracteres são maiúsculo
entrada.isupper()
```

```python
# verifica se realmente foi escrito algo ou é só um espaço em branco
entrada.isspace()
```

```python
apenas_espaco = input("Aperte apenas a barra de espaço: ")
apenas_espaco.isspace()
```

Também podemos fazer a verificação se a string tiver realmente algo escrito convertendo ela pro tipo boleano - `bool(variavel)` - onde irá retornar true se tiver algo e false se estiver vazia.

```python
booleano = bool(apenas_espaco)

booleano
```

```python
string_vazia = ""

booleano_falso = bool(string_vazia)

booleano_falso
```

Porém, o vazia quer dizer que a string realmente não possui nada escrito, por exemplo: **string = ""**

Caso ela possua espaço, a variável ao ser convertida para o tipo booleano irá retornar true porque irá considerar os caracteres que são apenas espaços em branco.

## Exercício 8

Faça um programa que receba os seguintes dados de um funcionário: nome, idade e salario. Na empresa que esse funcionário trabalha, seu salário é aumentado de ano em ano em R$ 800,90.

Sabendo disso, imprima o nome, idade e salário do funcionário daqui a 1 ano no seguinte formato: "O funcionário \<nome\>, daqui a 1 ano, terá \<idade\> anos, recebendo um salário igual a R$\<salario\>.

```python
nome = input("Digite seu nome: ")
idade = int(input("Digite sua idade: "))
salario = float(input("Digite o valor do seu salário: "))

valor_aumento = 800.90
```

```python
print(f"O funcionário {nome}, daqui a 1 ano, terá {idade + 1} anos, recebendo um salário igual a R$ {salario + valor_aumento}")
```
