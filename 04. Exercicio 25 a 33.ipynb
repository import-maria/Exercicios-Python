## Exercício 26

Faça um programa que verifique se um número é divisível por 3 e por 5 ao mesmo tempo.

```python
numero1 = float(input("Digite um número qualquer: "))
```

```python
if numero1 % 3 == 0 and numero1 % 5 == 0:
  print(f"O número {numero1} é divisivel por 5 e 3")

elif numero1 % 3 == 0 and not numero1 % 5 == 0:
  print(f"O número {numero1} é divisivel por 3")

elif numero1 % 5 == 0 and not numero1 % 3 == 0:
  print(f"O número {numero1} é divisivel por 5")

else:
  print(f"O número não é divisivel por 5 e nem por 3")
```

## Exercício 27

Faça um programa que calcule a média de três notas e exiba a situação do aluno (aprovado - maior ou igual a 70 -, em recuperação - entre 60 e 70 - ou reprovado - menor que 60).

```python
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))

media = (n1+n2+n3)/3
print(f"Média: {media}")

if media >= 70:
    print("Aprovado")
elif media >= 60:
    print("Em recuperação")
else:
    print("Reprovado")
```

## Exercício 29

Faça um programa que calcule o IMC (Índice de Massa Corporal) e exiba a categoria correspondente (abaixo do peso, peso normal, sobrepeso, obesidade grau I, obesidade grau II ou obesidade grau III).

<img src="https://endocrinologiacuritiba.com.br/wp-content/uploads/2012/02/tabela_imc.jpg" class="center">

```python
peso = float(input("Digite seu peso: "))
altura = float(input("Digite sua altura (em cm): "))
```

```python
IMC = (peso)/(altura/100) ** 2
print(f"IMC: {IMC}")
```

```python
if IMC < 18.5:
  print("Abaixo do peso normal")
elif IMC <= 24.9:
  print("Peso normal")
elif IMC <= 29.9:
  print("Excesso de Peso")
elif IMC <= 34.9:
  print("Obesidade classe I")
elif IMC <= 39.9:
  print("Obesidade classe II")
else:
  print("Obesidade classe III")
```

## Exercício 31

Faça um programa que receba o salário de um funcionário e calcule o valor do seu aumento de acordo com as seguintes condições: salários até R\$ 1.000, aumento de 20%; salários entre R\$ 1.000 e R\$ 3.000, aumento de 15%; salários acima de R\$ 3.000, aumento de 10%.

```python
salario = float(input("Digite seu salário: "))
```

```python
if salario <= 1.000:
  print(f"Seu salário com aumento de 20%: R$ {salario + (salario * 20)/100:,.2f}")
if salario <= 3.000:
  print(f"Seu salário com aumento de 15%: R$ {salario + (salario * 15)/100:,.2f}")
else:
  print(f"Seu salário com aumento de 10%: R$ {salario + (salario * 10)/100:,.2f}")
```

## Exercício 33

Faça um programa que receba a temperatura em graus Celsius e converta para Fahrenheit ou vice-versa, de acordo com a escolha do usuário.

```python
escolha = input("Você quer converter de Fahrenheit para Celsius? [y/n]: ").strip().lower()
temperatura = float(input("Digite a temperatura: "))

if escolha == 'y':
    c = 5*(temperatura - 32)/9
    print(f"A temperatura em Celsius é: {c:.2f}ºC")
else:
    f = (9/5)*temperatura + 32
    print(f"A temperatura em Fahrenheit é: {f:.2f}ºF")
```
