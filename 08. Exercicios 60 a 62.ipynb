## Exercício 60
Crie um programa que solicite ao usuário que insira um número. Utilize tratamento de exceções para garantir que o programa não quebre se o usuário inserir algo que não seja um número. Se ocorrer uma exceção, imprima uma mensagem informando que houve um erro.

```python
try:
  numero = int(input("Digite um número: "))
except ValueError as error:
  print("Houve um erro.")
  print(error)
else:
  print(numero)
```

## Exercício 61

Crie uma função que recebe dois números como parâmetros e realiza alguma operação. No programa principal, chame essa função e utilize um bloco try e except para capturar exceções. No bloco except, inclua uma cláusula para lidar com exceções de divisão por zero e outra cláusula para capturar qualquer outra exceção. Exiba mensagens adequadas para cada situação.

```python
divide = lambda n1, n2: n1/n2
```

```python
try:
  n1 = int(input("Digite o dividendo: "))
  n2 = int(input("Digite o divisor: "))
  divisao = divide(n1, n2)
except ZeroDivisionError:
  print("O divisor não pode ser zero!")
except Exception as error:
  print("Houve um erro!")
  print(error)
else:
  print(f"O resultado é: {divisao}")
```

## Exercício 62

Crie uma função chamada divisao_personalizada que recebe dois números como parâmetros. Se o segundo número for zero, a função deve gerar uma exceção personalizada chamada DivisaoPorZeroException. No programa principal, chame essa função e capture a exceção, exibindo uma mensagem adequada.

```python
class DivisaoPorZeroException(Exception):
  pass
```

```python
def divisao_personalizada(n1: int, n2: int):
  if n2 == 0:
    raise DivisaoPorZeroException("Você não pode dividir por zero!")
  return n1/n2
```

```python
try:
  n1 = int(input("Digite o dividendo: "))
  n2 = int(input("Digite o divisor: "))
  divisao = divisao_personalizada(n1, n2)
except ZeroDivisionError:
  print("O divisor não pode ser zero!")
except DivisaoPorZeroException as error:
  print(error)
else:
  print(f"O resultado é: {divisao}")
```
