Crie um programa que modele um animal de estimação:

- Desenvolva uma classe chamada AnimalDeEstimacao. A classe deve ter os atributos nome, especie e idade.
- Implemente métodos para permitir que o animal de estimação emita um som e se movimente.
- Crie uma instância da classe e demonstre o uso de seus métodos.

```python
class AnimalDeEstimacao:

  def __init__(self, nome, barulho, idade):
    self.nome = nome
    self.barulho = barulho
    self.idade = idade

  def emitir_som(self):
    print(f"O/A {self.nome} faz esse barulho: {self.barulho}!!!!")

  def movimentar(self):
    print(f"{self.nome} está começou a andar!")
```


## Exercício 64

Desenvolva um sistema que gerencie produtos em um estoque:

Implemente uma classe Produto com atributos como nome, preço e quantidade_em_estoque.
Crie métodos para adicionar e remover produtos do estoque.
Faça um pequeno programa que crie várias instâncias de Produto e demonstre a gestão do estoque.

```python
class Produto:

  def __init__(self, nome, preco, quantidade_em_estoque):
    self.nome = nome
    self.preco = preco
    self.quantidade_em_estoque = quantidade_em_estoque

  def adicionar_produto_no_estoque(self, quantidade):
    self.quantidade_em_estoque += quantidade

  def remover_produto_no_estoque(self, quantidade):
    self.quantidade_em_estoque -= quantidade

  def alterar_preco(self, novo_preco):
    self.preco = novo_preco

  def aplicar_desconto(self, percentual_de_desconto):
    self.preco -= (self.preco * percentual_de_desconto) / 100

    print(f"APROVEITEEE! {self.nome} está com {percentual_de_desconto}% de desconto!!")

  def consultar_estoque(self):
    print("----------------------------------------------")
    print(f"Produto:  {self.nome:}")
    print(f"Preço: R$ {self.preco}")
    print(f"Estoque: {self.quantidade_em_estoque} disponíveis")
    print("----------------------------------------------")
```

```python
batata = Produto("Batata", 10, 1000)
```

```python
batata.consultar_estoque()
```

```python
batata.remover_produto_no_estoque(300)
```

```python
batata.consultar_estoque()
```

```python
batata.adicionar_produto_no_estoque(100)
```

```python
batata.consultar_estoque()
```

```python
batata.aplicar_desconto(25)
```

```python
batata.consultar_estoque()
```

## Exercício 65

Crie um programa que simule um sistema bancário:

- Desenvolva uma classe ContaBancaria com atributos como numero_da_conta, titular e saldo.
- Implemente métodos para depositar, sacar e transferir dinheiro entre contas.
- Crie um pequeno sistema que permita criar novas contas e realizar operações bancárias entre elas.

```python
class Conta_Bancaria:

  def __init__(self, numero_da_conta, titular, saldo):
    self.__numero_da_conta = numero_da_conta
    self.titular = titular
    self.__saldo = saldo

  def depositar(self, quantidade):
    self.__saldo += quantidade

  def sacar(self, quantidade):
    if quantidade <= self.__saldo:
      self.__saldo -= quantidade
    else:
      print("Saldo insuficiente!")

  def transferencia(self, quanto_vai_transferir, destino):
    if quanto_vai_transferir <= self.__saldo:
      self.__saldo -= quanto_vai_transferir
      destino.__saldo += quanto_vai_transferir
    else:
      print("Saldo insuficiente!")

  def consultadr_dados_bancarios(self):
    print('----------------------------------------------')
    print(f"Número da conta: {self.__numero_da_conta}")
    print(f"Titular: {self.titular}")
    print(f"Saldo: {self.__saldo:,.2f}")
    print('----------------------------------------------')
```

```python
Maria = Conta_Bancaria('12345', "Maria Eduarda de Lima", 30000)
Laissa = Conta_Bancaria('12346', "Laissa Rosaura de Lima", 45000)

Maria.consultadr_dados_bancarios();
Laissa.consultadr_dados_bancarios()
```

```python
Maria.depositar(300)
Maria.consultadr_dados_bancarios();
```

```python
Laissa.transferencia(1500, Maria)
```

```python
Maria.consultadr_dados_bancarios();
Laissa.consultadr_dados_bancarios()
```
