## Exercício 67

Crie uma matriz 3x3 com valores aleatórios entre 1 e 10. Em seguida, calcule a média dos valores em cada linha e imprima o resultado.

```python
matriz = np.random.rand(3, 3) * 10

matriz
```

```python
# fixar os números aleatórios
np.random.seed(0)
array = np.random.randint(1, 10, size = (3, 3))
array
```

```python
# calculando a média de cada linha
array.mean(axis = 1)
```

```python
array.mean(axis = 0) # média das colunas
```

Utilizando as funções do pandas:

```python
matriz = pd.DataFrame(data = {'linha 1': [1, 2, 3],
                              'linha 2': [10, 7, 7],
                              'linha 3': [5, 6, 9]})
```

```python
matriz.describe()
```

```python
print(f"Média da coluna 'linha 1': {matriz['linha 1'].mean()}")
print(f"Média da coluna 'linha 2': {matriz['linha 2'].mean()}")
print(f"Média da coluna 'linha 3': {matriz['linha 3'].mean()}")
```

## Exercício 70
Crie um array unidimensional com 100 números inteiros aleatórios. Encontre o valor máximo, o valor mínimo, a média e o desvio padrão desses números e imprima os resultados.

```python
np.random.seed(0)

array_100 = np.random.randint(low=18, high=90, size=100)
array_100
```

```python
print(f"Valor mínimo: {array_100.min()}")
print(f"Valor máximo: {array_100.max()}")
print(f"Média: {array_100.mean()}")
print(f"Desvio Padrão: {array_100.std()}")
```

## Exercício 71
Suponha que você tenha um modelo que retorna probabilidades contínuas e você deseja aplicar um threshold para atribuir classes binárias (0 ou 1). Crie um programa que simule esse processo usando a função `np.where()` do NumPy.

A lista de probabilidades é a seguinte: `y_pred_probs = np.array([0.4, 0.6, 0.2. 0.3, 0.8, 0.9, 0.1, 0.8, 0.4, 0.5])`

```python
y_pred_probs = np.array([0.4, 0.6, 0.2, 0.3, 0.8, 0.9, 0.1, 0.8, 0.4, 0.5])
y_pred_probs
```

**O que é threshold?**

Threshold, em inglês, significa "limiar" ou "limite". Em diferentes contextos, o termo pode ter significados diferentes, mas geralmente se refere a um ponto crítico ou valor a partir do qual algo acontece ou é tomada uma decisão.

Ou seja, para esse exercicios iremos alterar os elementos dentro do array em relação a um limite, onde números menores que 0.5 serão alterados para zero e números maiores que o.5 serão alterados para 1.


```python
np.where(y_pred_probs >= 0.5, 1, 0)
```
